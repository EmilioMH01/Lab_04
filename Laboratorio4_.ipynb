## <font color= #6E6E6E> 5. Modelado de Microestructura </font>

Con la información obtenida del OB y OHLC, estimamos el Effective Spread de Roll utilizando la ecuación $ 2\surd| Cov(\triangle P_t,\triangle P_{t-1}|) $. 

```python
#Effective Spread BITCOIN/Binance
spread = pd.DataFrame(columns = ['timestamp', 'Close', 'Spread', 'EffectiveSpread'])
spread['timestamp'] = main.data_binance['datetime']
spread['Close'] = main.data_binance['Midprice']
spread['Spread'] = np.abs(main.data_binance['bidVolume'] - main.data_binance['askVolume'])
for i in range(data.rango-5):
    effective = np.sqrt(np.abs(np.cov([spread['Close'][i],spread['Close'][i+1]], [spread['Close'][i+4],spread['Close'][i+5]]).sum()))
    spread['EffectiveSpread'][i+5] = effective
spread.fillna(0)
```
